```python
import pandas as pd
import os

DATA_PATH = '../data/raw/phishing_site_urls.csv'
PROCESSED_DATA_DIR = '../data/processed'
os.makedirs(PROCESSED_DATA_DIR, exist_ok=True)

df = pd.read_csv(DATA_PATH)
```

```python
print("Thông tin tổng quan về bộ dữ liệu:")
print(df.info())
df.dropna(subset=['URL'], inplace=True)
print(f"\nSố dòng trước khi xóa trùng lặp: {len(df)}")
df.drop_duplicates(subset=['URL'], inplace=True)
print(f"Số dòng sau khi xóa trùng lặp: {len(df)}")
```

```python
df.rename(columns={'URL': 'url', 'Label': 'label_text'}, inplace=True)
df['label'] = df['label_text'].map({'bad': 1, 'good': 0})
df.drop(columns=['label_text'], inplace=True)
print("\nPhân bổ nhãn sau khi chuyển đổi:")
print(df['label'].value_counts())
```

```python
df_processed = df.sample(frac=1, random_state=42).reset_index(drop=True)
processed_path = os.path.join(PROCESSED_DATA_DIR, 'dataset_processed.csv')
df_processed.to_csv(processed_path, index=False)
print(f"\nHoàn tất! Dữ liệu sạch đã được lưu tại: {processed_path}")
```
